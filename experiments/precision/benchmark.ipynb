```python
import pywt
import numpy as np
```

```python
signal = np.random.rand(1000)
```

```python
wavelet = 'haar'

diff = pywt.idwt(*pywt.dwt(signal, wavelet), wavelet) - signal
```

```python
print(f"Max: {np.max(np.abs(diff))}")
print(f"Mean: {np.mean(np.abs(diff))}")
print(f"Std: {np.std(np.abs(diff))}")
print(f"Energy: {np.sum(diff**2)}")
```
