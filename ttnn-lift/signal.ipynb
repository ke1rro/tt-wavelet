# Signal Types

```python
from tt_wavelet.signal import Signal, SignalType, InputSpec
import matplotlib.pyplot as plt
import numpy as np
```

```python
sig_normal_spec = InputSpec(
    kind=SignalType.NORMAL,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_normal = Signal(sig_normal_spec)

sig_sinusoidal_spec = InputSpec(
    kind=SignalType.SINUSOIDAL,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_sinusoidal = Signal(sig_sinusoidal_spec)

sig_alternating_spec = InputSpec(
    kind=SignalType.ALTERNATING,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_alternating = Signal(sig_alternating_spec)

sig_impulse_spec = InputSpec(
    kind=SignalType.IMPULSE,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_impulse = Signal(sig_impulse_spec)

sig_ramp_spec = InputSpec(
    kind=SignalType.RAMP,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_ramp = Signal(sig_ramp_spec)

sig_uniform_spec = InputSpec(
    kind=SignalType.UNIFORM,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_uniform = Signal(sig_uniform_spec)

sig_constant_spec = InputSpec(
    kind=SignalType.CONSTANT,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_constant = Signal(sig_constant_spec)

sig_step_spec = InputSpec(
    kind=SignalType.STEP,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_step = Signal(sig_step_spec)

sig_square_spec = InputSpec(
    kind=SignalType.SQUARE,
    length=100,
    magnitude=1.0,
    seed=1,
)
sig_square = Signal(sig_square_spec)
```

## Visual Representation of Signal Types

```python
generated_signals = [
    ("Normal Signal", sig_normal),
    ("Sinusoidal Signal", sig_sinusoidal),
    ("Alternating Signal", sig_alternating),
    ("Impulse Signal", sig_impulse),
    ("Ramp Signal", sig_ramp),
    ("Uniform Signal", sig_uniform),
    ("Constant Signal", sig_constant),
    ("Step Signal", sig_step),
    ("Square Signal", sig_square),
]

plt.figure(figsize=(14, 10))
for idx, (title, signal) in enumerate(generated_signals, start=1):
    plt.subplot(3, 3, idx)
    plt.title(title)
    plt.plot(signal.data, marker="o")

plt.tight_layout()
plt.show()
```
